# Nested Loops

```python
# Examples -> unique pairs

for i in range(1,5):
  for j in range(1,5):
    print(i,j)
```

```python
# code here

rows = int(input('enter number of rows'))

for i in range(1,rows+1):
  for j in range(1,i+1):
    print(j,end='')
  print()
```

### Pattern 2

```python
# Code here
rows = int(input('enter number of rows'))

for i in range(1,rows+1):
  for j in range(1,i+1):
    print(j,end='')
  for k in range(i-1,0,-1):
    print(k,end='')

  print()
```

# Loop Control Statement

- Break
- Continue
- Pass

```python
for i in range(1,10):
  if i == 5:
    break
  print(i)
```

```python
lower = int(input('enter lower range'))
upper = int(input('enter upper range'))

for i in range(lower,upper+1):
  for j in range(2,i):
    if i%j == 0:
      break
  else:
    print(i)
```

```python
# Continue
for i in range(1,10):
  if i == 5:
    continue
  print(i)
```

```python
for i in range(1,10):
  pass
```
